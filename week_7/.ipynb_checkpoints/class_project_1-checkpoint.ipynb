# GOOGLE PLAY

```python
# to display first 7 rows
import pandas as pd

#retrieve the csv file
data = pd.read_csv("Top-Apps-in-Google-Play.csv")


for i in range(0,7):
    df = data.iloc[i]
    # print excel
    print(df)
    print()
```

```python
#first 3 columns
import pandas as pd

data = pd.read_csv("Top-Apps-in-Google-Play.csv")


columns = list(data)

#add first 3 items in columns to a new list
stuff = []
for i in range(0,3):
    stuff.append(columns[i])
print(stuff)

for j in stuff:
    col = (data[j])
    print(col)
    print()
```
