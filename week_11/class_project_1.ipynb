```python
import pandas as pd

#data['Age']


class Students:
    
    data = pd.read_csv("sis.csv")
    pirates = dict(data)
    thing = {}
    
    #pirate
    mat_no = []
    name = []
    age = []
    grade = []
    

    
    
    
    def set_group(self):
        
         
        for i in (self.data.index):
            j = self.data.loc[i]["Age"]
            if j > 14 and j < 18:
                #self.thing[(self.data.loc[i]["Name"])] = self.data.loc[i]["Age"]
                self.sort(i)
                self.pir_lyst = {'Mat_No':self.mat_no,'Name':self.name, 'Age': self.age, 'Grade':self.grade}
                
        self.mat_no = []
        self.name = []
        self.age = []
        self.grade = []
        
        for i in (self.data.index):
            j = self.data.loc[i]["Age"]
            if j > 18 and j <22:
                self.sort(i)
                self.yan_lyst = {'Mat_No':self.mat_no,'Name':self.name, 'Age': self.age, 'Grade':self.grade}

                
        self.mat_no = []
        self.name = []
        self.age = []
        self.grade = []
        
        for i in (self.data.index):
            j = self.data.loc[i]["Age"]
            if j > 22:
                self.sort(i)
                self.bul_lyst = {'Mat_No':self.mat_no,'Name':self.name, 'Age': self.age, 'Grade':self.grade}
        
    def sort(self,i):
            self.mat_no.append(self.data.loc[i]['MatNo'])
            self.name.append(self.data.loc[i]['Name'])
            self.age.append(self.data.loc[i]['Age'])
            self.grade.append(self.data.loc[i]['Grade'])
                
                
    def pirate(self):
                
        pir = pd.DataFrame(self.pir_lyst)
        print(pir)
        pir.to_csv('Pirates.csv')
        
    def yankees(self):
                
        yan = pd.DataFrame(self.yan_lyst)
        print(yan)
        yan.to_csv('Yankees.csv')
        
    def bulls(self):
                
        bul = pd.DataFrame(self.bul_lyst)
        print(bul)
        bul.to_csv('Bulls.csv')
    
wow = Students()

wow.set_group()
choice = input("Enter p for pirates, y for yankees, b for bulls: ")

if choice == "p":
    wow.pirate()
elif choice == "y":
    wow.yankees()
elif choice == "b":
    wow.bulls()
else:
    print("We have not created that feature yet")
    
```
